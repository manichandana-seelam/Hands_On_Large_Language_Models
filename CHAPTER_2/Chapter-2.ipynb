```python
# Import the tokenizer automatically suitable for Model
from transformers import AutoTokenizer


# This is simply a list of background colors.
#Each token will be displayed in a different color.

colors_list = [
    '102;194;165',
    '252;141;98',
    '141;160;203',
    '231;138;195',
    '166;216;84',
    '255;217;47'
]

# A function with two arguments
   # sentence = input text ,  tokenizer name
def show_tokens(sentence, tokenizer_name):

  # downloads and loads whichever tokenizer you specify.
    tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)

  # tokenizer splits the text into tokens
  # tokens into token ids
  # stores only input ids
    token_ids = tokenizer(sentence).input_ids

 # enumerate gives the token ids with their position
 # idx  = the index position of the token
 # t  = token ids
    for idx, t in enumerate(token_ids):
        print(

            #  x1b = Ascii escape character it tells to start formatting text
            #  0;30;48;2  = This controls the style. ( 0=Reset formatting style after a token was generated with color , 30= text color , 48=background color , 2 : Set the background color using RGB values)
            #  Pick a background color based on the current token position.
            #  m = Apply the formatting
            f'\x1b[0;30;48;2;{colors_list[idx % len(colors_list)]}m'

            # Converts token ids back to text
            + tokenizer.decode(t)

            # End of formatting
            + '\x1b[0m',
            end=' '
        )

# Input text
text = """
English and CAPITALIZATION
show_tokens False None elif == >= else: two tabs:" " Three tabs:
"   "
12.0*50=600

"""
# Functiona call with input and tokenizer name
show_tokens(text, "bert-base-uncased")
```

```python
from transformers import AutoModel, AutoTokenizer

# Load a tokenizer
tokenizer = AutoTokenizer.from_pretrained("microsoft/deberta-base")

# Load a language model
model = AutoModel.from_pretrained("microsoft/deberta-v3-xsmall")

# Tokenize the sentence
tokens = tokenizer('Hello world', return_tensors='pt')

# Process the tokens
# input_ids → token numbers
# ** for attention_mask → tells the model which positions are real tokens
# sends token IDs and attention mask into the DeBERTa model
# gets the final contextual embedding vector for every token.
output = model(**tokens)[0]

# shows how many sentences, how many tokens, and how many numbers represent each token.
print(output.shape)

# Raw output of a language model
print(output)
```

```python
# The tokenizer split your input into tokens, including any special tokens
# tokens['input_ids'][0]:  it is the no.of input sentences in this case we have only one so 0
for token in tokens['input_ids'][0]:

  #decoding the token ids
  print(tokenizer.decode(token))
```

```python
# Pandas is a Python library used for Reading data
import pandas as pd

# urllib  = Python library for working with URLs
# Inside urllib there is a module called request. It knows how to open websites
from urllib import request


# Go to request module and open and download that link
data = request.urlopen('https://storage.googleapis.com/maps-premium/dataset/yes_complete/train.txt')


# read the file and now you get raw bytes
# converts bytes into text (utf-8) is the encoding format.
# split into lines (song by song) Every line becomes one list element.
# Line 0 and 1 contains metadata (details like id , name , artist) so ignore it and starts from 2

lines = data.read().decode("utf-8").split('\n')[2:]

# Only keep playlists having more than one song.
# Remove playlists with only one song
playlists = [s.rstrip().split() for s in lines if len(s.split())
> 1]

# Load song metadata
# Download the song information file from the internet and store it in songs_file.
songs_file = request.urlopen('https://storage.googleapis.com/maps-premium/dataset/yes_complete/song_hash.txt')


# Reads the downloaded file, converts it into readable text, and splits it into separate lines, storing each line as an element in a list.
songs_file = songs_file.read().decode("utf-8").split('\n')


# Removes extra whitespace from each line, s
# plits each line into id, title, and artist using the tab (\t) character,
# stores the result as a list of lists
songs = [s.rstrip().split('\t') for s in songs_file]


#Creates a Pandas DataFrame from the songs list and names the columns as id, title, and artist.
songs_df = pd.DataFrame(data=songs, columns = ['id', 'title', 'artist'])


# Sets the id column as the index of the DataFrame so each song can be identified by its ID.
songs_df = songs_df.set_index('id')


# Prints the playlist, with each playlist containing a list of song IDs.
print( 'Playlist #1:\n ', playlists[0], '\n')
print( 'Playlist #2:\n ', playlists[1])
```

```python
# import word2vec model
from gensim.models import Word2Vec

# Train our Word2Vec model
# Each playlist acts like a sentence.
# Each song ID acts like a word.
# vector size =Every song gets a vector of 32 numbers.
# window = It looks the 20 neighbour songs from right and left
# For every positive example, generate 50 negative examples for training model
# min_count=1  keep the songs that appears one time it means kepts every song
# workers = use 4 cpu cores for faster training
model = Word2Vec(playlists, vector_size=32, window=20, negative=50, min_count=1, workers=4)

song_id = 2172

# wv = wordvectors that have embeddings
# most_similar = finds the vector that closes to given vector
# finding the most similar songs to that song id
# word2vec stores the song id in string format
model.wv.most_similar(positive=str(song_id))

print(songs_df.iloc[2172])
```

```python
# This function finds the 5 songs that are most similar to the given song_id using the trained Word2Vec model and gets their IDs.
 # It then looks up those song IDs in the songs_df table and returns their titles and artists as recommendations.
import numpy as np

# Creating a function with one argument
def print_recommendations(song_id):
  similar_songs = np.array(model.wv.most_similar(positive=str(song_id),topn=5))[:,0]
  return  songs_df.iloc[similar_songs]
# Extract recommendations
print_recommendations(2172)
```
