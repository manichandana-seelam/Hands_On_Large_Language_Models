```python
 # transformer = Hugging face python library that provides the transformer based models

 #                                  ===IMPORTING===

 # AutoModelForCausalLM = Automatically selects and loads the suitable causal LM
 # AutoTokenizer = Automatically loads the suitable suitable tokenizer for the loaded model
 # pipeline = creates an interace for using the model
 # torch = transformers uses the PyTorch as the backend for running the models

 from transformers import AutoModelForCausalLM , AutoTokenizer
 import torch


#  Model name on Hugging face you want to load
model_name="microsoft/Phi-3-mini-4k-instruct"



# Automatically loads the pretrained model
model=AutoModelForCausalLM.from_pretrained(
    model_name,

    device_map="auto",         # Loads the model on gpu/cpu which is available (For gpu use = "cuda"  , cpu="cpu")

    torch_dtype="auto",        # Automatically selects the datatype for computation

    trust_remote_code=True,    # Allows the custom code to run if needed

)

# Load the tokenizer that automatically converts the text into tokens and generated tokens into back to text
tokenizer=AutoTokenizer.from_pretrained(model_name)



```

```python
# Importing the pipeline which combines the model and tokenizer for text generation

from transformers import pipeline

generator=pipeline(
    "text-generation",               # Tells the transformers to generate the text
    model=model,
    tokenizer=tokenizer,             # Previously loaded tokenizer and model
    return_full_text=False,          # Output comntains only newly generated tokens not the input ones
    max_new_tokens=500,              # MAx no.of. new tokens to generate
    do_sample=False                  # This tells how the next token is generated (Always chooses the highest probability)   TRUE = Randomly chooses

)
```

```python
# Chat style input It has 2 keys "role" = Who is speaking    ,  content = message send by user

messages=[
    { "role":"user" , "content" : "Create a funny joke about chickens."}
]


# Pipeline internally does Converts chat messages into the model's expected prompt format,Tokenizes the prompt , Runs the model  , Generates new tokens , Converts tokens back into text.
output=generator(messages)

# Model generates a list of output select the first generated output
print(output[0]["generated_text"])
```
